1. Compute the camera calibration matrix and distortion coefficients given a set of chessboard images.
2. Apply a distortion correction to raw images.
3. Use color transforms, gradients, etc., to create a thresholded binary image.
4. Apply a perspective transform to rectify binary image ("birds-eye view").
5. Detect lane pixels and fit to find the lane boundary.

8. Output visual display of the lane boundaries and numerical estimation of lane curvature and vehicle position.

# 3. Use color transforms, gradients, etc., to create a thresholded binary image.

Lets define different transforms that we can apply on undistorted images produced above to get thresholded binary images. Remember to use mpimg.imread outputs in RGB while cv2.imread in BGR (great for confusing). Here after experimenting with HLS channels and sobel gradients, combination of white/yellow masks leads to good results.

```python
from utils import threshold

bin_thresholded = threshold.threshold(test_image)
plt.imshow(bin_thresholded, cmap='gray')
plt.title('Thresholded Binary')
```

# 4. Apply a perspective transform to rectify binary image ("birds-eye view").

The results of some of the perspective transforms don't look very encouraging as of now. But after many tries this seems to be the norm. Because the lane lines on one side are dotted & almost non-existent many binary images just show one line. Frame with shadows is also bringing noise although I thought use of L (light) channel will help me reduce the effect of shadows. Now the hope is that the line fit algorithms with reuse of earlier fits will help fill the gaps in lane detection and smooth things out.

```python
binary_warped, M, Minv = camera.perspective_transform(bin_thresholded)
plt.imshow(binary_warped, cmap='gray')
plt.title("Bird's eye view")
```

# 5. Detect lane pixels and fit to find the lane boundary.

### Using sliding windows & fit reuse
After unsuccessful attempt with windows centroids, let's try sliding window with fit reuse.
Find Lane points using histogram, separate non-zero points from binary image, for every window pick the points which are inside window and get mean position of points to form the next window. Repeat step to get left/right fits and fit a 2nd order polynomial curve on lanes.
Reuse left/right fits for next frames from the above excercise done on first frame.

```python
from utils import fit_lane

# HYPERPARAMETERS
window_height = 80 # Break image into 9 vertical layers since image height is 720
# Choose the number of sliding windows 720 (frame height)/80 (window height) = 9
nwindows = 9
# Set the width of the windows +/- margin
margin = 100
# Set minimum number of pixels found to recenter window
minpix = 50

leftx, lefty, rightx, righty, out_img = fit_lane.find_lane_pixels(binary_warped, nwindows, margin, minpix)
left_fit, right_fit, left_fitx, right_fitx, ploty = fit_lane.fit_poly(binary_warped.shape, leftx, lefty, rightx, righty)

# print(left_fit, right_fit)

out_img[lefty, leftx] = [255, 0, 0]
out_img[righty, rightx] = [0, 0, 255]


plt.figure(figsize=(10,40))
plt.subplot(1,2,1)
plt.imshow(binary_warped, cmap='gray')
plt.title('Binary Warped')

plt.subplot(1,2,2)
plt.imshow(out_img)
plt.plot(left_fitx, ploty, color='yellow')
plt.plot(right_fitx, ploty, color='yellow')
plt.xlim(0, 1280)
plt.ylim(720, 0)
plt.title('Lane Detected')
```

# 6. Determine the curvature of the lane and vehicle position with respect to center.
## Radius Of Curvature and Position from center
Calculating values and putting it on the image

```python
ym_per_pix = 3.0/72.0 # meters per pixel in y dimension
xm_per_pix = 3.7/660.0 # meters per pixel in x dimension
y_eval = 700
midx = 640

# left_fit = [  1.74344564e-04 , -2.55400146e-01 ,  4.48433182e+02]
# right_fit = [  4.43840644e-04 , -6.43052083e-01 ,  1.28913810e+03]

y1 = (2*left_fit[0]*y_eval + left_fit[1])*xm_per_pix/ym_per_pix
y2 = 2*left_fit[0]*xm_per_pix/(ym_per_pix*ym_per_pix)

curvature = ((1 + y1*y1)**(1.5))/np.absolute(y2)
print("Radius of Curvature: %f" % curvature)
# print(left_fit)
# print(right_fit)

x_left_pix = left_fit[0]*(y_eval**2) + left_fit[1]*y_eval + left_fit[2]
x_right_pix = right_fit[0]*(y_eval**2) + right_fit[1]*y_eval + right_fit[2]
position_from_center = ((x_left_pix + x_right_pix)/2 - midx) * xm_per_pix
if position_from_center < 0:
    text = 'left'
else:
    text = 'right'
print('Distance From Center: %.2fm %s' % (np.absolute(position_from_center), text))
```

# 7. Warp the detected lane boundaries back onto the original image.
reuse the left/right fits, second iteration

```python
result = fit_lane.find_lane_pixels_with_prev_fit2(undistorted_image, binary_warped, left_fit, right_fit, Minv)

plt.imshow(result)
```

# 8. Output visual display of the lane boundaries and numerical estimation of lane curvature and vehicle position.
### Putting everything on the image

```python
cv2.putText(result,'Radius of Curvature: %.2fm' % curvature,(20,40), cv2.FONT_HERSHEY_SIMPLEX, 1,(255,255,255),2)
cv2.putText(result,'Distance From Center: %.2fm %s' % (np.absolute(position_from_center), text),(20,80), cv2.FONT_HERSHEY_SIMPLEX, 1,(255,255,255),2)
plt.imshow(result)
```
